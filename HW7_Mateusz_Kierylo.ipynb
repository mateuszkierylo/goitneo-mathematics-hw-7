### **Tasks on determining all possible variants**


**6)** Three dice are thrown. What is more likely: to get a total of 11 or 12 points?

```python
import numpy as np
import pandas as pd

# Create all possible outcomes for three dice
dice_outcomes = np.arange(1, 7)
all_combinations_df = pd.DataFrame(
    np.array(np.meshgrid(dice_outcomes, dice_outcomes, dice_outcomes)).T.reshape(-1, 3),
    columns=['Die1', 'Die2', 'Die3']
)

print(all_combinations_df)

# Calculate the sum of each combination
all_combinations_df['Sum'] = all_combinations_df.sum(axis=1)

# Count combinations that sum to 11 and 12
count_11 = all_combinations_df[all_combinations_df['Sum'] == 11].shape[0]
count_12 = all_combinations_df[all_combinations_df['Sum'] == 12].shape[0]

# Total number of combinations
total_combinations = all_combinations_df.shape[0]

# Calculate probabilities
probability_11 = count_11 / total_combinations
probability_12 = count_12 / total_combinations

print("\nthe probability of rolling 11 points is:", probability_11)
print("\nthe probability of rolling 12 points is:", probability_12)



```

**7)** On ten identical cards, numbers from 0 to 9 are written. Determine the probability that a randomly formed number using these cards

a) is a two-digit number multiple of 18

b) is a three-digit number multiple of 36

**RESOLVE:**

a) For a number to be divisible by 18, it must be divisible by 2 and 9 at the same time:
* Divisible by 2: the number ends with an even digit (0, 2, 4, 6, 8).
* Divisibility by 9: the sum of the digits of the number is a multiple of 9.

Zrzut ekranu 2024-05-7 o 19.40.27.png


```python
# Define a function to check if a number is a multiple of 18
def is_multiple_of_18(n):
    return n % 18 == 0

# Create a range of two-digit numbers from 10 to 99
two_digit_numbers = range(10, 100)

# Filter the list of numbers to find those that are multiples of 18
multiples_of_18 = [num for num in two_digit_numbers if is_multiple_of_18(num)]

# Calculate the probability

probability = len(multiples_of_18) / len(two_digit_numbers)

# Print the list of numbers divisible by 18 and the probability
print(f"Numbers divisible by 18: {multiples_of_18}")
print(f"Probability that a randomly formed two-digit number is a multiple of 18: {probability:.5f}")
```

b) A number divisible by 36 must also be divisible by 4 and 9.

* Divisibility by 4: the last two digits of the number must form a number divisible by 4.

* Divisibility by 9: the sum of the digits of the number must be a multiple of 9.

Zrzut ekranu 2024-05-7 o 19.40.37.png

```python
# Define a function to check if a number is a multiple of 36
def is_multiple_of_36(num):
    if num % 4 != 0:
        return False
    return sum(int(digit) for digit in str(num)) % 9 == 0

# Range of three-digit numbers from 100 to 999
three_digit_numbers = range(100, 1000)
# List of three-digit numbers that are divisible by 36
multiples_of_36 = [num for num in three_digit_numbers if is_multiple_of_36(num)]

# Calculate the probability of a randomly chosen three-digit number being divisible by 36
probability = len(multiples_of_36) / len(three_digit_numbers)
# Print the calculated probability
print(f"Numbers divisible by 36: {multiples_of_36}")
print(f"Probability: {probability:.5f}")
```
